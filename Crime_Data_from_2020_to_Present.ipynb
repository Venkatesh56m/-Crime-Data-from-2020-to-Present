```python
df.rename(columns={
    'DR_NO': 'Report Number',
    'Date Rptd': 'Date Reported',
    'DATE OCC': 'Date Occurred',
    'TIME OCC': 'Time Occurred',
    'AREA': 'Area Code',
    'AREA NAME': 'Area Name',
    'Rpt Dist No': 'Reporting District',
    'Part 1-2': 'Crime Category Part',
    'Crm Cd': 'Primary Crime Code',
    'Crm Cd Desc': 'Primary Crime Description',
    'Mocodes': 'MO Codes',
    'Vict Age': 'Victim Age',
    'Vict Sex': 'Victim Sex',
    'Vict Descent': 'Victim Descent',
    'Premis Cd': 'Premise Code',
    'Premis Desc': 'Premise Description',
    'Weapon Used Cd': 'Weapon Code',
    'Weapon Desc': 'Weapon Description',
    'Status': 'Status Code',
    'Status Desc': 'Status Description',
    'Crm Cd 1': 'Crime Code 1',
    'Crm Cd 2': 'Crime Code 2',
    'Crm Cd 3': 'Crime Code 3',
    'Crm Cd 4': 'Crime Code 4',
    'LOCATION': 'Location Description',
    'Cross Street': 'Cross Street',
    'LAT': 'Latitude',
    'LON': 'Longitude'
}, inplace=True)
```

```python
df
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_').str.replace('-','_')
```

```python
df.columns
```

```python
columns_to_keep = ['report_number', 'date_reported', 'date_occurred', 'time_occurred',
                  'area_name','crime_code_1', 'crime_code_2',
       'crime_code_3', 'crime_code_4','status_code', 'status_description','location_description', 'cross_street',
       'latitude', 'longitude']
```

```python
columns_to_keep
```

```python
df = df[columns_to_keep]
```

```python
df['date_reported'] = pd.to_datetime(df['date_reported'], errors ='coerce')
df['date_occurred'] = pd.to_datetime(df['date_occurred'], errors = 'coerce')
```

```python
df.isnull().sum()
```

```python
missing_percentage = (df.isnull().sum()/len(df) * 100)
```

```python
missing_percentage = missing_percentage.sort_values(ascending=False)
```

```python
missing_percentage
```

```python
df.drop(columns = ['crime_code_4','crime_code_3','crime_code_2','cross_street'],inplace=True)
```

```python
df.columns
```

```python
# extracting the usefull time features 

df['Year'] = df['date_occurred'].dt.year
df['Month'] = df['date_occurred'].dt.month
df['Day'] = df['date_occurred'].dt.day
df['Hour'] = df['time_occurred'] // 100
```

```python
df
```

```python
print(df['date_occurred'])
print(df['date_reported'])
print(df['time_occurred'])
```

```python
df['date_occurred'] = pd.to_datetime(df['date_occurred']).dt.strftime('%d-%m-%Y')
df['date_reported'] = pd.to_datetime(df['date_reported']).dt.strftime('%d-%m-%Y')
```

```python
# Pad with zeros (in case any value is like '930' → '0930')
df['time_occurred'] = df['time_occurred'].astype(str).str.zfill(4)

# Convert to HH:MM format
df['time_occurred'] = df['time_occurred'].str.slice(0, 2) + ':' + df['time_occurred'].str.slice(2, 4)
```
